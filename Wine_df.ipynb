```python
wine_df = pd.read_csv(file,  encoding="ISO-8859-2")
wine_df.head()
```

```python
wine_ratings_df = pd.DataFrame(wine_df)
wine_ratings_df.head()
```

```python
wine_avg_score=wine_ratings_df.groupby(['points']).mean()
wine_score_clean=wine_avg_score.sort_index(ascending=False)

wine_pricetoscore_df = pd.DataFrame(wine_score_clean['price'])
wine_pricetoscore_df
```

```python
ranked_df=wine_ratings_df.groupby(['winery']).mean()
ranked_df1=pd.DataFrame(ranked_df)
results=ranked_df1.sort_values(['points'], ascending=False)
best_winery_df=results.drop('First numerical', 1)
best_winery_df
```
